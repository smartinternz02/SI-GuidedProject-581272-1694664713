# Importing dataset and Preprocessing

```python
data.drop(["EmployeeNumber"],axis=True,inplace=True)
```

```python
data.drop(["StandardHours","Over18","EmployeeCount"],axis=True,inplace=True)
```

```python
columns=data.select_dtypes(include=np.number).columns.tolist()
columns
```

```python
for i in columns:
    plt.figure()
    sns.boxplot(data=data,x=i)
    plt.title(i)
    plt.show()
    
```

```python
corelation=data.corr(numeric_only=True)
corelation
```

```python
plt.subplots(figsize=(20,10))
sns.heatmap(corelation,annot=True)
```

```python
x = data.drop(columns=["Attrition"])
y = data["Attrition"]
print(x)
print(y)
print(x.shape)
print(y.shape)
print(type(x))
print(type(y))
```

```python
x.info()
```

```python
y.info()
```

```python
# label encoding
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
le
```

```python
columns=x.select_dtypes(include=object).columns.tolist()
columns
```

```python
for i in columns:
    x[i]=le.fit_transform(x[i])
```

```python
y=le.fit_transform(y)
```

```python
x
```

```python
# feature scalling
from sklearn.preprocessing import MinMaxScaler
ms=MinMaxScaler()
x_scaled=pd.DataFrame(ms.fit_transform(x),columns=x.columns)
x_scaled.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.1,random_state=0)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

# Logestic Regression

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr
```

```python
lr.fit(x_train,y_train)
```

```python
lr_pred=lr.predict(x_test)
lr_pred
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
dtc
```

```python
dtc.fit(x_train,y_train)
```

```python
dtc_pred=dtc.predict(x_test)
dtc_pred
```

```python
# Hyper parameter tuning
from sklearn.model_selection import GridSearchCV
parameter={
'criterion':['gini','entropy'],
  'splitter':['best','random'],
  'max_depth':[1,2,3,4,5],
  'max_features':['auto', 'sqrt', 'log2']}
```

```python
grid_search=GridSearchCV(estimator=dtc,param_grid=parameter,cv=5,scoring="accuracy")
```

```python
grid_search
```

```python
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_params_
```

```python
dtc_cv=DecisionTreeClassifier(criterion= 'gini',
 max_depth= 4,
 max_features= 'sqrt',
 splitter= 'random')
dtc_cv.fit(x_train,y_train)
```

```python
dtc_pred=dtc_cv.predict(x_test)
```

```python
dtc_pred
```

# Evaluation

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,roc_auc_score,roc_curve
```

```python
accuracy_score(y_test,lr_pred)
```

```python
accuracy_score(y_test,dtc_pred)
```

```python
confusion_matrix(y_test,lr_pred)
```

```python
confusion_matrix(y_test,dtc_pred)
```

```python
pd.crosstab(y_test,lr_pred)
```

```python
pd.crosstab(y_test,dtc_pred)
```

```python
print(classification_report(y_test,lr_pred))
```

```python
print(classification_report(y_test,dtc_pred))
```

```python
dtc_prob=dtc.predict_proba(x_test)[:,1]
dtc_prob
```

```python
lr_prob=lr.predict_proba(x_test)[:,1]
lr_prob
```

```python
lr_fpr,lr_tpr,lr_threshsholds=roc_curve(y_test,lr_prob,)
dtc_fpr,dtc_tpr,dtc_threshsholds=roc_curve(y_test,dtc_prob)
```

```python
plt.plot(dtc_fpr,dtc_tpr)
plt.xlabel("fpr")
plt.ylabel("tpr")
plt.title("roc curve Logestic Regression")
plt.show()
```

```python
plt.plot(dtc_fpr,dtc_tpr)
plt.xlabel("fpr")
plt.ylabel("tpr")
plt.title("roc curve Decision Tree")
plt.show()
```

```python
plt.plot(dtc_fpr, dtc_tpr, label='Decision Tree Classifier')
plt.plot(lr_fpr, lr_tpr, label='Logistic Regression')
plt.legend()
plt.show()
```
